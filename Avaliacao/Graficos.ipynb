```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('avaliacao.csv',sep=',')
dados
```

```python
colunas = dados.columns.values[1:]


for i in range(len(colunas)):
    fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(16,16))
    c = colunas[i]
    count = dados[c].groupby(dados[c].values).count()
    axs.pie(count.values, labels=count.index.values, autopct='%1.0f%%')
    axs.set_title(c)
    plt.savefig(str(i+1)+'.jpg')
    fig.clf()
    axs.cla()
    plt.close()
```
