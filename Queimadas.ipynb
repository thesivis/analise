```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
total = pd.read_csv('total.csv', sep=';')
total
```

```python
total_ate_agosto = 105508
```

```python
media_por_anos = []

for i in range(len(total)):
    media_por_anos.append(total['Total'][-1*(i+1):].mean())
    
media_por_anos = list(reversed(media_por_anos))
```

```python
fig = plt.figure(figsize=(20,10),facecolor='white')
ax = plt.axes()


font = {'family': 'serif',
        'weight': 'normal',
        'size': 20,
        }

x = list(range(16,0,-1))
ax.plot( media_por_anos);

ax.plot([total_ate_agosto]*(len(x)))

locs, labels = plt.xticks()
plt.xticks(list(range(0,16)), x, fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Quantidade de anos considerada', fontdict=font)
plt.title('Média do total de queimadas até dia 31/08 de cada ano', fontdict=font)


plt.savefig('queimadas.jpg')
```
